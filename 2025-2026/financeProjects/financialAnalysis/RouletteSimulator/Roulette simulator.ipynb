```python
roulette = ['z']
for i in range(1, 37):
    if i % 2 == 0:
        roulette.append('b')
    else:
        roulette.append('r')

print(roulette)
```

```python
import random

looping = 1000;
statistics = {}

def roulette_martingale(statistics, nb_round):
    nb_rounds = 1000
    z_counts = 0
    r_counts = 0
    b_counts = 0

    player_bet = 'b'
    fund = 100000
    initial_fund = fund
    bet_amount = 10
    reset_amount = bet_amount
    lose_flag = False
    bankrupt_flag = False

    for i in range(nb_rounds):
        if fund <= 0:
            print(f"nb_round {nb_round}: Game over! Let's sell the house...")
            statistics[nb_round] = {"status": "bankrupt", "amount": 0}
            bankrupt_flag = True
            break

        if lose_flag:
            bet_amount = min(bet_amount * 2, fund)

        outcome = random.randint(0, 36)

        if roulette[outcome] == 'z':
            z_counts += 1
        elif roulette[outcome] == 'r':
            r_counts += 1
        elif roulette[outcome] == 'b':
            b_counts += 1

        if roulette[outcome] == player_bet:
            fund += bet_amount
            lose_flag = False
            bet_amount = reset_amount
        else:
            fund -= bet_amount
            lose_flag = True

    print(f"Black: {b_counts}, Red: {r_counts}, Zero: {z_counts}")
    print(f"Final fund: {fund}")

    if bankrupt_flag:
        return 

    if fund > initial_fund:
        print(f"Profit: +{fund - initial_fund}")
        statistics[nb_round] = {"status": "profit", "amount": fund - initial_fund}
    elif fund < initial_fund:
        print(f"Loss: -{initial_fund - fund}")
        statistics[nb_round] = {"status": "loss", "amount": initial_fund - fund}


def roulette_inverse_martingale(statistics, nb_round):
    nb_rounds = 1000
    z_counts = 0
    r_counts = 0
    b_counts = 0

    player_bet = 'b'
    fund = 100000
    initial_fund = fund
    bet_amount = 10  # Initial bet amount
    reset_amount = bet_amount
    win_flag = False  # Tracks whether the last bet was won
    bankrupt_flag = False

    for i in range(nb_rounds):
        if fund <= 0:
            print(f"nb_round {nb_round}: Game over! Let's sell the house...")
            statistics[nb_round] = {"status": "bankrupt", "amount": 0}
            bankrupt_flag = True
            break

        if win_flag:
            bet_amount = min(bet_amount * 2, fund)  # Double bet only after a win
        else:
            bet_amount = reset_amount  # Reset bet after a loss

        outcome = random.randint(0, 36)

        if roulette[outcome] == 'z':
            z_counts += 1
        elif roulette[outcome] == 'r':
            r_counts += 1
        elif roulette[outcome] == 'b':
            b_counts += 1

        if roulette[outcome] == player_bet:
            fund += bet_amount
            win_flag = True  # Mark this round as a win
        else:
            fund -= bet_amount
            win_flag = False  # Mark this round as a loss

    print(f"Black: {b_counts}, Red: {r_counts}, Zero: {z_counts}")
    print(f"Final fund: {fund}")

    if bankrupt_flag:
        return 

    if fund > initial_fund:
        print(f"Profit: +{fund - initial_fund}")
        statistics[nb_round] = {"status": "profit", "amount": fund - initial_fund}
    elif fund < initial_fund:
        print(f"Loss: -{initial_fund - fund}")
        statistics[nb_round] = {"status": "loss", "amount": initial_fund - fund}
```

```python
for i in range(looping) : 
    roulette_martingale(statistics,i)
```

```python

profit_values = [result["amount"] for result in statistics.values() if result["status"] == "profit"]
loss_values = [result["amount"] for result in statistics.values() if result["status"] in ["loss", "bankrupt"]]

max_profit = max(profit_values, default=0)
total_rounds = len(statistics)
profit_percentage = (len(profit_values) / total_rounds) * 100 if total_rounds > 0 else 0
loss_percentage = (len(loss_values) / total_rounds) * 100 if total_rounds > 0 else 0
bankrupt_count = sum(1 for result in statistics.values() if result["status"] == "bankrupt")
bankrupt_percentage = (bankrupt_count / total_rounds) * 100 if total_rounds > 0 else 0
average_profit = sum(profit_values) / len(profit_values) if profit_values else 0
average_loss = sum(loss_values) / len(loss_values) if loss_values else 0

print(f"Maximum Profit: {max_profit}")
print(f"Percentage of Profitable Rounds: {profit_percentage:.2f}%")
print(f"Percentage of Losing Rounds (Including Bankruptcies): {loss_percentage:.2f}%")
print(f"Number of Bankruptcies: {bankrupt_count}")
print(f"Percentage of Bankruptcies: {bankrupt_percentage:.2f}%")
print(f"Average Winning Amount: {average_profit}")
print(f"Average Losing Amount (Including Bankrupt Cases): {average_loss}")
```
