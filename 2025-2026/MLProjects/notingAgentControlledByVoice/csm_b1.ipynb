```python
import torch
from transformers import CsmForConditionalGeneration, AutoProcessor
```

```python
model_id = "models/csm-1b"
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# load the model and the processor
processor = AutoProcessor.from_pretrained(model_id)
model = CsmForConditionalGeneration.from_pretrained(model_id, device_map=device)
```

```python
conversation = [
    {"role": "0", "content": [{"type": "text", "text": "Bonjour, C'est Iasted, comment je peux vous aider ?"}]},
]
inputs = processor.apply_chat_template(
    conversation,
    tokenize=True,
    return_dict=True,
).to(device)
```

```python
audio = model.generate(
    **inputs,
    output_audio=True,
    max_new_tokens=2000,   # increase this
    min_new_tokens=400,    # optional: forces at least some length
    do_sample=False,
    depth_decoder_do_sample=False,
)
processor.save_audio(audio, "french_example.wav")
```
